```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a figure and set up two subplots (1 row, 2 columns)
f, axs = plt.subplots(1, 2, figsize=(12, 6))

# Scatter plot 1: Quantity vs Sales for Region A
sns.scatterplot(data=data[data['Region'] == 'A'], x="Quantity", y="Sales", hue="Region", ax=axs[0])
axs[0].set_title('Region A: Quantity vs Sales')

# Scatter plot 2: Quantity vs Sales for Region B
sns.scatterplot(data=data[data['Region'] == 'B'], x="Quantity", y="Product", hue="Region", ax=axs[1])
axs[1].set_title('Region B: Quantity vs Sales')

# Adjust layout to prevent overlapping
f.tight_layout()

# Show the plots
plt.show()
```

```python
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True)#cmap="RdBu"
sns.set_style("whitegrid")
sns.set_palette("viridis")
plt.title("Feature Correlation Heatmap")
plt.show()
```

### Creating Practice

```python
sns.set(style="whitegrid")

# Plotting
plt.figure(figsize=(10, 6))

# Boxplot to show distribution of Sales across Regions
sns.boxplot(data=data, x="Region", y="Sales", color="b")

# Add titles and labels
plt.title("Sales Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
# Encode 'Region' to numeric
data['Region_encoded'] = pd.factorize(data['Quantity'])[0]

# Plotting with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x="Region_encoded", y="Sales", color="b")

# Add titles and labels
plt.title("Relationship between Quantity and Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")

plt.show()
```
