```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(42)

# Create a sample DataFrame
data = {
    'age': [23, 45, np.nan, 23, 34, 56, 29, np.nan, 33, 41],
    'salary': [50000, 60000, 70000, 48000, 52000, np.nan, 58000, 61000, np.nan, 64000],
    'department': ['HR', 'Engineering', 'HR', 'Engineering', 'HR', 'Engineering', np.nan, 'HR', 'Engineering', 'HR'],
    'start_date': ['2018-01-01', '2019-02-15', '2020-06-01', '2018-09-22', '2017-07-18', '2015-03-30', '2021-02-18', '2019-06-05', '2020-11-30', '2016-12-25']
}

df = pd.DataFrame(data)

# Convert start_date to datetime format
df['start_date'] = pd.to_datetime(df['start_date'])

# Display the initial data
df
```

```python
# Inspecting the DataFrame
print(df.info())  # General info (data types, non-null counts)
print(df.describe())  # Summary statistics

# Checking for missing values
print(df.isnull().sum())
```

```python
# Drop rows with any missing values
df_cleaned = df.dropna()

# Drop rows with missing values in specific columns
df_cleaned_salary = df.dropna(subset=['salary'])

# Drop columns that have any missing values
df_cleaned_columns = df.dropna(axis=1)

# Display the cleaned dataframe
print(df_cleaned)
```

```python
# Impute missing 'salary' with the mean salary
df['salary'] = df['salary'].fillna(df['salary'].mean())

# Impute missing 'age' with the median age
df['age'] = df['age'].fillna(df['age'].median())

# Impute missing 'department' with the mode (most frequent value)
df['department'] = df['department'].fillna(df['department'].mode()[0])

print(df)
```

```python
# Check for duplicates
duplicates = df.duplicated()
print(duplicates)

# Drop duplicate rows
df_cleaned_no_duplicates = df.drop_duplicates()

# Display cleaned data
print(df_cleaned_no_duplicates)
```

```python
# Boxplot to visualize outliers in 'salary' and 'age'
plt.figure(figsize=(12, 6))

# Plot salary
plt.subplot(1, 2, 1)
sns.boxplot(x=df['salary'])
plt.title('Salary Boxplot')

# Plot age
plt.subplot(1, 2, 2)
sns.boxplot(x=df['age'])
plt.title('Age Boxplot')

plt.tight_layout()
plt.show()
```

```python
# Calculate IQR for salary and age
Q1 = df[['salary', 'age']].quantile(0.25)
Q3 = df[['salary', 'age']].quantile(0.75)
IQR = Q3 - Q1

# Define outlier conditions for salary and age
outlier_condition = ((df[['salary', 'age']] < (Q1 - 1.5 * IQR)) | (df[['salary', 'age']] > (Q3 + 1.5 * IQR)))

# Remove rows with outliers
df_no_outliers = df[~outlier_condition.any(axis=1)]

# Display cleaned data
print(df_no_outliers)
```

```python
# Convert 'salary' to integer type (if it's not already)
df['salary'] = df['salary'].astype(int)

# Convert 'department' to categorical type
df['department'] = df['department'].astype('category')

# Display data types after conversion
print(df.dtypes)
```

```python
# Extract year and month from 'start_date'
df['start_year'] = df['start_date'].dt.year
df['start_month'] = df['start_date'].dt.month

# Display the new features
print(df[['start_date', 'start_year', 'start_month']])
```

```python
# Visualize the cleaned data with pairplot
sns.pairplot(df[['age', 'salary', 'start_year']])
plt.show()

# Visualize salary distribution
sns.histplot(df['salary'], kde=True, bins=10)
plt.title('Salary Distribution')
plt.show()
```
